```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
y = [1, 2, 3,4]
plt.plot(y)
plt.ylabel('some numbers')
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [1, 3, 6, 9]
plt.plot(x, y)
plt.show()
```

```python
t = np.arange(0, 5, 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
t = np.arange(0, 5, 0.2)
l1 = plt.plot(t, t, 'r--', label='t')
l2 = plt.plot(t, t**2, 'bs', label='t**2')
l3 = plt.plot(t, t**3, 'g^', label='t**3')
plt.legend()
plt.show()
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.title('scatter')
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
names = ['a', 'b', 'c']
values = [1, 10, 100]

plt.figure(1, figsize=(9,3))

plt.subplot(131)
plt.xlabel('bar')
plt.bar(names, values)

plt.subplot(132)
plt.xlabel('scatter')
plt.scatter(names, values)

plt.subplot(133)
plt.xlabel('plot')
plt.plot(names, values)

plt.suptitle('Categorical Plotting')
plt.show()
```

```python
a = np.random.random(1000)
b = np.random.randn(1000)

plt.figure(1, figsize=(12, 3))

plt.subplot(121)
plt.title('random')
plt.hist(a, 20)
plt.grid(True)

plt.subplot(122)
plt.title('randn')
plt.hist(b, 20)
plt.grid(True)

plt.show()
```
